```python
#velocidades
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Cargar datos ──────────────────────────────────────────────────────────────
data_298 = np.loadtxt('veloc_298.xvg', comments=['#', '@'])
data_400 = np.loadtxt('veloc_400.xvg', comments=['#', '@'])

tiempo_298 = data_298[:, 0]
tiempo_400 = data_400[:, 0]

indices = [4, 8, 12, 16, 20]
colores = ['steelblue', 'tomato', 'seagreen', 'darkorange', 'mediumpurple']
nombres = ['Átomo 1', 'Átomo 2', 'Átomo 3', 'Átomo 4', 'Átomo 5']

# ── Función de suavizado (media móvil) ────────────────────────────────────────
def suavizar(y, ventana=50):
    return np.convolve(y, np.ones(ventana)/ventana, mode='same')

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 7), sharex=True)
fig.suptitle('Velocidades de 5 átomos — 298 K vs 400 K', fontsize=13, fontweight='bold')

# ── 298 K ─────────────────────────────────────────────────────────────────────
for idx, color, nombre in zip(indices, colores, nombres):
    raw = data_298[:, idx]
    ax1.plot(tiempo_298, raw, linewidth=0.4, alpha=0.2, color=color)
    ax1.plot(tiempo_298, suavizar(raw), linewidth=1.8, color=color, label=nombre)

ax1.set_ylabel('|V| (nm/ps)', fontsize=11)
ax1.set_title('298 K', fontsize=11)
ax1.legend(fontsize=9, ncol=5, loc='upper right')

# ── 400 K ─────────────────────────────────────────────────────────────────────
for idx, color, nombre in zip(indices, colores, nombres):
    raw = data_400[:, idx]
    ax2.plot(tiempo_400, raw, linewidth=0.4, alpha=0.2, color=color)
    ax2.plot(tiempo_400, suavizar(raw), linewidth=1.8, color=color, label=nombre)

ax2.set_ylabel('|V| (nm/ps)', fontsize=11)
ax2.set_xlabel('Tiempo (ps)', fontsize=11)
ax2.set_title('400 K', fontsize=11)
ax2.legend(fontsize=9, ncol=5, loc='upper right')

plt.tight_layout()
plt.savefig('velocidades_comparacion.png', dpi=300, bbox_inches='tight')
print("Guardado!")

# ── Estadísticas ──────────────────────────────────────────────────────────────
print("\n── Velocidad media por átomo (nm/ps) ─────────────────────────────")
print(f"{'':12}  {'298 K':>10}  {'400 K':>10}  {'std 298K':>10}  {'std 400K':>10}")
for idx, nombre in zip(indices, nombres):
    print(f"{nombre:12}  {np.mean(data_298[:, idx]):>10.4f}  "
          f"{np.mean(data_400[:, idx]):>10.4f}  "
          f"{np.std(data_298[:, idx]):>10.4f}  "
          f"{np.std(data_400[:, idx]):>10.4f}")
```

```python
#diedros
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# ── Cargar datos ──────────────────────────────────────────────────────────────
arg_298 = np.loadtxt('arg-3.dat')
arg_400 = np.loadtxt('arg-3-400.dat')

phi_298 = arg_298[:, 1];  psi_298 = arg_298[:, 2]
phi_400 = arg_400[:, 1];  psi_400 = arg_400[:, 2]

tiempo_298 = np.linspace(0, 2, len(phi_298))
tiempo_400 = np.linspace(0, 2, len(phi_400))

# ── Figura ────────────────────────────────────────────────────────────────────
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Ángulos diedros — ARG-3', fontsize=14, fontweight='bold')

# ── Panel izquierdo: phi ──────────────────────────────────────────────────────
ax1.plot(tiempo_298, phi_298, color='steelblue', linewidth=0.8, alpha=0.9, label='298 K')
ax1.plot(tiempo_400, phi_400, color='tomato',    linewidth=0.8, alpha=0.9, label='400 K')
ax1.set_xlabel('Tiempo (ps)', fontsize=11)
ax1.set_ylabel('φ (°)', fontsize=12)
ax1.set_title('Evolución temporal φ', fontsize=11)
ax1.set_ylim(-180, 180)
ax1.axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax1.legend(fontsize=9)

# ── Panel central: psi ───────────────────────────────────────────────────────
ax2.plot(tiempo_298, psi_298, color='steelblue', linewidth=0.8, alpha=0.9, label='298 K')
ax2.plot(tiempo_400, psi_400, color='tomato',    linewidth=0.8, alpha=0.9, label='400 K')
ax2.set_xlabel('Tiempo (ps)', fontsize=11)
ax2.set_ylabel('ψ (°)', fontsize=12)
ax2.set_title('Evolución temporal ψ', fontsize=11)
ax2.set_ylim(-180, 180)
ax2.axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax2.legend(fontsize=9)

# ── Panel derecho: Ramachandran con densidad KDE ──────────────────────────────
# densidad de color para 298K
xy_298 = np.vstack([phi_298, psi_298])
kde_298 = gaussian_kde(xy_298)(xy_298)
idx_298 = kde_298.argsort()
ax3.scatter(phi_298[idx_298], psi_298[idx_298], c=kde_298[idx_298],
            cmap='Blues', s=8, alpha=0.7, label='298 K')

# densidad de color para 400K
xy_400 = np.vstack([phi_400, psi_400])
kde_400 = gaussian_kde(xy_400)(xy_400)
idx_400 = kde_400.argsort()
ax3.scatter(phi_400[idx_400], psi_400[idx_400], c=kde_400[idx_400],
            cmap='Reds', s=8, alpha=0.7, label='400 K')

# líneas de referencia
ax3.axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax3.axvline(0, color='gray', linewidth=0.5, linestyle='--')

# etiquetas de regiones bien colocadas según los datos reales
# 298K está arriba izquierda → región beta
# 400K está abajo → zona poliprolina/extendida
mean_phi_298 = np.mean(phi_298)
mean_psi_298 = np.mean(psi_298)
mean_phi_400 = np.mean(phi_400)
mean_psi_400 = np.mean(psi_400)

ax3.annotate(f'β\n({mean_phi_298:.0f}°, {mean_psi_298:.0f}°)',
             xy=(mean_phi_298, mean_psi_298), xytext=(mean_phi_298+15, mean_psi_298-25),
             fontsize=9, color='navy', fontweight='bold',
             arrowprops=dict(arrowstyle='->', color='navy', lw=1))

ax3.annotate(f'400K\n({mean_phi_400:.0f}°, {mean_psi_400:.0f}°)',
             xy=(mean_phi_400, mean_psi_400), xytext=(mean_phi_400+15, mean_psi_400-25),
             fontsize=9, color='darkred', fontweight='bold',
             arrowprops=dict(arrowstyle='->', color='darkred', lw=1))

ax3.set_xlabel('φ (°)', fontsize=12)
ax3.set_ylabel('ψ (°)', fontsize=12)
ax3.set_title('Ramachandran', fontsize=11)
ax3.set_xlim(-180, 180)
ax3.set_ylim(-180, 180)

# leyenda manual
from matplotlib.lines import Line2D
leyenda = [Line2D([0],[0], marker='o', color='w', markerfacecolor='steelblue', markersize=7, label='298 K'),
           Line2D([0],[0], marker='o', color='w', markerfacecolor='tomato',    markersize=7, label='400 K')]
ax3.legend(handles=leyenda, fontsize=9)

plt.tight_layout()
plt.savefig('diedros_arg3.png', dpi=300, bbox_inches='tight')
print("Guardado!")

print("\n── ARG-3 ─────────────────────────────────────────────────────────")
print(f"  phi 298K:  media = {np.mean(phi_298):.1f}°  |  std = {np.std(phi_298):.1f}°")
print(f"  phi 400K:  media = {np.mean(phi_400):.1f}°  |  std = {np.std(phi_400):.1f}°")
print(f"  psi 298K:  media = {np.mean(psi_298):.1f}°  |  std = {np.std(psi_298):.1f}°")
print(f"  psi 400K:  media = {np.mean(psi_400):.1f}°  |  std = {np.std(psi_400):.1f}°")
```
