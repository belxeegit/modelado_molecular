```python
#ramachandran
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
 
# ── Cargar datos ──────────────────────────────────────────────────────────────
ala = np.loadtxt('ala-2.dat')
arg = np.loadtxt('arg-3.dat')
leu = np.loadtxt('leu-4.dat')
 
residuos = [
    (ala, 'ALA-2', 'Blues',  'steelblue'),
    (arg, 'ARG-3', 'Reds',   'tomato'),
    (leu, 'LEU-4', 'Greens', 'seagreen'),
]
 
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Diagramas de Ramachandran — Simulación extendida 500 ps (298 K)',
             fontsize=13, fontweight='bold')
 
for ax, (data, nombre, cmap, color) in zip(axes, residuos):
    phi = data[:, 1]
    psi = data[:, 2]
 
    # densidad KDE
    xy  = np.vstack([phi, psi])
    kde = gaussian_kde(xy)(xy)
    idx = kde.argsort()
    sc  = ax.scatter(phi[idx], psi[idx], c=kde[idx], cmap=cmap,
                     s=4, alpha=0.7)
    plt.colorbar(sc, ax=ax, label='Densidad')
 
    # líneas de referencia
    ax.axhline(0, color='gray', linewidth=0.5, linestyle='--')
    ax.axvline(0, color='gray', linewidth=0.5, linestyle='--')
 
    # etiquetas de regiones — alejadas de la nube (arriba izquierda)
    ax.text(-170, 170,  r'$\beta$',    fontsize=13, color='darkred',   fontweight='bold')
    ax.text(-60,  -40,  r'$\alpha_R$', fontsize=13, color='navy',      fontweight='bold')
    ax.text(80,   40,   r'$\alpha_L$', fontsize=13, color='darkgreen', fontweight='bold')
 
    ax.set_xlim(-180, 180)
    ax.set_ylim(-180, 180)
    ax.set_xlabel('φ (°)', fontsize=12)
    ax.set_ylabel('ψ (°)', fontsize=12)
    ax.set_title(nombre, fontsize=12, fontweight='bold')
 
    # cajita de estadísticas abajo a la derecha, lejos de la nube
    ax.text(0.98, 0.02,
            f'φ: {np.mean(phi):.1f}° ± {np.std(phi):.1f}°\nψ: {np.mean(psi):.1f}° ± {np.std(psi):.1f}°',
            transform=ax.transAxes, fontsize=8,
            verticalalignment='bottom', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
 
plt.tight_layout()
plt.savefig('ramachandran_opcional.png', dpi=300, bbox_inches='tight')
print("Guardado: ramachandran_opcional.png")
 
print("\n── Estadísticas ──────────────────────────────────────────────────")
for data, nombre, _, _ in residuos:
    phi = data[:, 1]; psi = data[:, 2]
    print(f"  {nombre}:  phi = {np.mean(phi):.1f}° ± {np.std(phi):.1f}°  |  psi = {np.mean(psi):.1f}° ± {np.std(psi):.1f}°")
```

```python
#histogramas
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# ── Cargar datos ──────────────────────────────────────────────────────────────
temp  = np.loadtxt('temp.xvg',  comments=['#', '@'])[:, 1]
veloc = np.loadtxt('veloc.xvg', comments=['#', '@'])
vx = veloc[:, 1]
vy = veloc[:, 2]
vz = veloc[:, 3]

# ── Figura 2x2 ────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Histogramas — Simulación extendida 500 ps (298 K)',
             fontsize=14, fontweight='bold')

datos = [
    (temp, 'Temperatura (K)',  'steelblue', 'navy',   'Temperatura'),
    (vx,   'Vx (nm/ps)',       'tomato',    'darkred', 'Componente Vx — Átomo 1'),
    (vy,   'Vy (nm/ps)',       'seagreen',  'darkgreen','Componente Vy — Átomo 1'),
    (vz,   'Vz (nm/ps)',       'darkorange','saddlebrown','Componente Vz — Átomo 1'),
]

for ax, (d, xlabel, color, lcolor, titulo) in zip(axes.flat, datos):
    mu, std = np.mean(d), np.std(d)

    ax.hist(d, bins=60, color=color, edgecolor='white', linewidth=0.4,
            alpha=0.8, density=True, label='Distribución simulada')

    x = np.linspace(mu - 4*std, mu + 4*std, 400)
    ax.plot(x, norm.pdf(x, mu, std), color=lcolor, linewidth=2,
            label=f'Gaussiana\n$\\mu$ = {mu:.2f}\n$\\sigma$ = {std:.2f}')

    ax.set_xlim(mu - 4*std, mu + 4*std)
    ax.set_xlabel(xlabel, fontsize=11)
    ax.set_ylabel('Densidad', fontsize=11)
    ax.set_title(titulo, fontsize=11)
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('histogramas_opcional.png', dpi=300, bbox_inches='tight')
print("Guardado: histogramas_opcional.png")

print("\n── Estadísticas ──────────────────────────────────────────────────")
print(f"  Temperatura:  mu = {np.mean(temp):.2f} K   |  std = {np.std(temp):.2f} K")
print(f"  Vx:           mu = {np.mean(vx):.4f} nm/ps |  std = {np.std(vx):.4f} nm/ps")
print(f"  Vy:           mu = {np.mean(vy):.4f} nm/ps |  std = {np.std(vy):.4f} nm/ps")
print(f"  Vz:           mu = {np.mean(vz):.4f} nm/ps |  std = {np.std(vz):.4f} nm/ps")
```
